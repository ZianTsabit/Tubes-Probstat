Membuat Visualisasi plot distribusi, dalam bentuk  histogram dan boxplot untuk setiap kolom numerik. Berikan uraian penjelasan kondisi setiap kolom berdasarkan kedua plot tersebut.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('water_potability.csv', index_col=0)
df.head()
```

```python
# Menyatakan kolom-kolom numerik

col_numeric = [
    'pH', 
    'Hardness', 
    'Solids', 
    'Chloramines', 
    'Sulfate', 
    'Conductivity', 
    'OrganicCarbon', 
    'Trihalomethanes', 
    'Turbidity'
]
```

```python
for col in col_numeric: 
    try:      
        df[col] = pd.to_numeric(df[col]) 
        df.hist(column=col)
    except ValueError:
        print('This column can not be represented as a histogram')
```

```python
for col in col_numeric: 
    plt.figure()
    df.boxplot([col])
```
